```python
path = '../input/world-population-live-dataset/World Population Live Dataset.csv'
df = pd.read_csv(path)
```

```python
df['CCA3'] = country_converter.convert(names=df['Name'], to="ISO3")
```

```python
px.choropleth(df, 
              locations = 'CCA3',
              color=df["2022"],
              color_continuous_scale ='dense',
              title='World population for 2022',
              height=800)

             
```

## the previous choropleth is fine, but we cant just maake 8 of them to show all the population per year for every county as this wouldn't be beneficial to compare, so will make a new dataset with diffrent characteristics to better show our work:

```python
df1 = (df.set_index(["Name","CCA3",'Area (km²)','Density (per km²)','GrowthRate','World Population Percentage','Rank'])
         .stack()
         .reset_index(name='Population')
         .rename(columns={'level_7':'Year'}))
```

```python
df1.Population = df1.Population*1000
```

```python
px.choropleth(df1.sort_values('Year'), 
              locations = 'CCA3',
              color="Population",
              animation_frame='Year',
              color_continuous_scale ='OrRd',
              title='World population through years' ,
              height=800)

             
```

```python
df2 = df1[df1['Population']<250000000]
px.choropleth(df2.sort_values('Year'), 
              locations = 'CCA3',
              color="Population",
              animation_frame='Year',
              color_continuous_scale ='OrRd',
              title='Africe population through years' ,
              scope = 'africa',
              height=800)

             
```

```python
px.choropleth(df1.sort_values('Year'), 
              locations = 'CCA3',
              color="Population",
              animation_frame='Year',
              color_continuous_scale ='OrRd',
              title='Asia population through years with China and India' ,
              scope = 'asia',
              height=800)

             
```
